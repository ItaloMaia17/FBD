### Acesso ao banco de dados e consultas

```python
import pandas as pd
import psycopg2 as pg
import sqlalchemy
from sqlalchemy import create_engine
```

```python
con = 'postgresql://postgres:Admin@localhost/STREAMING DE MUSICA'
sqlalchemy.create_engine (con)
```

```python
#consulta 1
query = 'Select us.nome, alb.titulo from usuario us, album alb where us.id= alb.id_criador'
result = pd.read_sql_query(query, con)
result
```

```python
#consulta 2
query2 = "select m.titulo as musica, gen.nome_genero as gênero from musica m join genero_da_musica gdm on m.id_musica = gdm.id_musica join genero_musical gen on gen.id_genero = gdm.id_genero order by m.titulo asc"
result2 = pd.read_sql_query(query2, con)
result2
```

```python
#consulta 3
query3 = "Select  gm.nome_genero, count(*) as total_musicas from musica m join genero_da_musica gdm on m.id_musica = gdm.id_musica join genero_musical gm on gm.id_genero = gdm.id_genero group by gm.nome_genero"
result3 = pd.read_sql_query(query3, con)
result3
```
